# Change Wallpaper

In this example we will download an image and use it as a allpaper.

## Initializing Automagica

The first step in building a script with Automagica is to import all the Automagica activities.

```python
from automagica import *
```

## Downloading the file

We will use a sample image file directly downloaded from the Automagica website. The image can be found and viewed on:

https://automagica.com/examples/wallpaper.jpg

If you would like to add your own variation, for example by looking up images from other resources on the web we recommend going over the browser automation examples and documentation. 

We can specify a path, but by default the download_file_from_url activity will download to the homedir.

```python
# Download the file 
downloaded_image = download_file_from_url('https://automagica.com/examples/wallpaper.jpg')
```

To see the full path to the newly downloaded image we can do so by simply running it's name

```python
downloaded_image
```

## Set the file as wallpaper

Last step is the set the newly downloaded image as wallpaper

```python
# Set image as wallpaper
set_wallpaper(downloaded_image)
```
