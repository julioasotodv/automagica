# Recorder example with Windows Calculator

This example shows how to use the Automagica Wand functionality to easily find elements on a screen. This eliminates the tedious task of finding the techinical identifier of elements.

In addition, sometimes working with the recorder is the only way to reliably find buttons and elements, for example when working with a virtualized desktop such as Citrix, TeamViewer, 

In this example we will automate the Windows Calculator by identifying the buttons using the Automagica Wand, clicking and finally reading the output with Automagica vision (ocr).

## Starting the recorder

The recorder can be opened by either pressing the Recorder button in the Automagica Lab, by running 'Automagica Wand' in a Windows console on a system with Automagica installed, or by running the following commands in a Python console:

```python
from automagica import *

recorder()
```

## Setting your target

The Automagica Wand relies on vision to identify elements on the screen. It uses a smart algorithm to find the best match which makes it robust to minor changes in the UI.

For starters, after choosing the preferred action in the recorder you have to identify your target by dragging a box around the element. Make sure you only select the element you want to target, including irrelevant changing elements in the target could cause a mismatch.

## Anchors

When your target is subject to change it is a good idea to pick an anchor. An anchor is an element that does not change in appearance and does not change in relative position to your target. 

For example, in browser automation when clicking on the first Google result, this result is subject to change depending on what you searched. However, there are a lot of elements that are static relative to the first result:
