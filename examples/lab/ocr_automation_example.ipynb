# Optical Character Recognition

In this example we will go over the Optical Character Recognition possibilities of Automagica

## Initializing Automagica

The first step in building a script with Automagica is to import all the Automagica activities.

```python
from automagica import *
```

## Downloading a sample file

To demonstrate the OCR capabilities we will use a sample image file directly downloaded from the Automagica website. The image can be found and viewed on:

https://automagica.com/examples/ocr_test.png

There is no text-layer on this jpg image, the goal is to read the text while using OCR.



We could specify a path, but by default the download_file_from_url activity will download to the homedir.

```python
# Download the file 
downloaded_image = download_file_from_url('https://automagica.com/examples/ocr_test.png')
```

We can now open the file to view it's contents 

```python
open_file(downloaded_image)
```

## Reading with OCR

Let's attempt to read the text of the newly downloaded image

```python
# Read image
extract_text_ocr(r"C:\Users\TvT\Desktop\ocr_test.png")
```

## Automate using OCR

Another use-case for this activity could be to automatically find elements like buttons on the screen and click on them, for example.

Make sure the image is open and visible on the screen, next we can find the coördinates of the word "underwear"

```python
find_text_on_screen_ocr('underwear')
```

We can also click on the word!

```python
click_on_text_ocr('underwear')
```
