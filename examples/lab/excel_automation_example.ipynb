## Saving Data

As a next step we can save the data and and exit Excel.

We first declare a path using the Automagica built-in home_path function to make sure we know where it is saved

```python
# Declare path to save it to, in this case in the homedir with the name "awesome_excel.xlsx"
excel_save_path = home_path('awesome_excel.xlsx')
```

```python
# Save the worksheet, by default this is in the homedir
excel.save_as(excel_save_path)
```

We can now exit Excel

```python
excel.quit()
```

As a last stap we can re-open the file we just saved to verify:

```python
open_file(excel_save_path)
```
